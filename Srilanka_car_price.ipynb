# Algorithm

```python
scaler_1 = MinMaxScaler()
```

```python
df_3[['Capacity','Mileage','Age']] = scaler_1.fit_transform(df_3[['Capacity','Mileage','Age']])
```

```python
df_3.head()
```

```python
df_3 = pd.get_dummies(df_3,['Brand','Model','Condition','Transmission','Body','Fuel','Seller_type'])
```

```python
df_3.info()
```

```python
X = df_3.drop(columns='Price')
```

```python
Y = df_3['Price']
```

```python
X_train_final,X_test_final,Y_train_final,Y_test_final = train_test_split(X,Y,test_size = 0.2,random_state=12)
```

```python
import xgboost as xgb
```

```python
xgb_model = xgb.XGBRegressor(n_estimators=2000, max_depth=15, eta=0.1, subsample=0.9, colsample_bytree=0.8)
```

```python
xgb_model.fit(X_train_final,Y_train_final)
```

```python
def mean_absolute_percentage_error(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
xgb_model.score(X_test_final,Y_test_final)
```

```python
xgb_predict = xgb_model.predict(X_test_final)
mean_absolute_percentage_error(Y_test_final, xgb_predict)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr = RandomForestRegressor(max_depth=10)
rfr.fit(X_train_final,Y_train_final)
```

```python
rfr.score(X_test_final,Y_test_final)
```

```python
rfr_predict = rfr.predict(X_test_final)
mean_absolute_percentage_error(Y_test_final, rfr_predict)
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR().fit(X_train_final,Y_train_final)
```
